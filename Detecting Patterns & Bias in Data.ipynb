```python
# Load dataset
file_path = "GenZ_DatingApp_Data.csv"  # Ensure the correct path
data = pd.read_csv('GenZ_DatingApp_Data.csv')

# Print column names to debug
print("Columns in dataset:", data.columns)

# Ensure 'Daily_Usage_Time' exists before processing
if 'Daily_Usage_Time' in data.columns:
    # Convert 'Daily_Usage_Time' to numerical values (hours)
    usage_mapping = {
        '30 minutes': 0.5, '1 hour': 1, '1.5 hours': 1.5, 
        '2 hours': 2, '2.5 hours': 2.5, '3 hours': 3
    }
    
    # Apply mapping and fill unknown values with NaN
    data['Daily_Usage_Time_Hours'] = data['Daily_Usage_Time'].map(usage_mapping)
else:
    print("Warning: 'Daily_Usage_Time' column not found in dataset.")

# Ensure 'Age' and 'Gender' exist before plotting
if 'Age' in data.columns and 'Gender' in data.columns:
    # Histogram: Age distribution across gender
    plt.figure(figsize=(10, 6))
    sns.histplot(data=data, x='Age', hue='Gender', multiple='stack', bins=10, kde=True)
    plt.title('Age Distribution Across Genders')
    plt.xlabel('Age')
    plt.ylabel('Count')
    plt.legend(title="Gender")
    plt.show()
else:
    print("Warning: 'Age' or 'Gender' column not found in dataset.")
```

```python
# Load dataset
file_path = "GenZ_DatingApp_Data.csv"  # Update the path if needed
data = pd.read_csv('GenZ_DatingApp_Data.csv')

# Ensure 'Daily_Usage_Time' column exists before processing
if 'Daily_Usage_Time' in data.columns:
    # Convert 'Daily_Usage_Time' to numerical values (hours)
    usage_mapping = {
        '30 minutes': 0.5, '1 hour': 1, '1.5 hours': 1.5, 
        '2 hours': 2, '2.5 hours': 2.5, '3 hours': 3
    }
    data['Daily_Usage_Time_Hours'] = data['Daily_Usage_Time'].map(usage_mapping)

# Check if required columns exist
if "Daily_Usage_Time_Hours" in data.columns and "Location" in data.columns:
    plt.figure(figsize=(10, 6))
    sns.kdeplot(data=data, x='Daily_Usage_Time_Hours', hue='Location', fill=True, common_norm=False)
    plt.title('Daily App Usage Time Across Locations')
    plt.xlabel('Daily Usage Time (Hours)')
    plt.ylabel('Density')
    plt.legend(title="Location")
    plt.show()
else:
    print("Warning: 'Daily_Usage_Time_Hours' or 'Location' column not found in dataset.")
```

```python
# Load dataset
file_path = "GenZ_DatingApp_Data.csv"  # Ensure correct file path
data = pd.read_csv('GenZ_DatingApp_Data.csv')

# Check if required columns exist
if {'Gender', 'Location', 'Age'}.issubset(data.columns):
    demographic_counts = {
        'Gender': data['Gender'].value_counts(),
        'Location': data['Location'].value_counts(),
        'Age Group': pd.cut(data['Age'], bins=[18, 22, 26, 30, 35, 40], right=False).value_counts().sort_index()
    }
    
    # Convert to DataFrame for better visualization
    demographic_df = pd.DataFrame(demographic_counts)
    print(demographic_df)
else:
    print("Warning: One or more required columns ('Gender', 'Location', 'Age') are missing from the dataset.")
```
